Vamos fazer um programa para verificar quem  o assassino de um crime.
Para descobrir o assassino, a polcia faz um pequeno questionrio com 5
perguntas onde a resposta s pode ser sim ou no:
a. Mora perto da vtima?
b. J trabalhou com a vtima?
c. Telefonou para a vtima?
d. Esteve no local do crime?
e. Devia para a vtima?
Cada resposta sim d um ponto para o suspeito. A polcia considera que os
suspeitos com 5 pontos so os assassinos, com 4 a 3 pontos so cmplices e
2 pontos so apenas suspeitos, necessitando outras investigaes. Valores
iguais ou abaixo de 1 so liberados.

```python
suma = 0
location = input('Do you live close to the victim: ')
if location == 'yes':
    suma = 1
work = input('Did you work with the victim: ')
if work == 'yes':
    suma += 1
call = input('Did you call to the victim: ')
if call == 'yes':
    suma += 1
crime_local = input('Did you go to the crime location: ')
if crime_local == 'yes':
    suma += 1
money = input('Did you owe money to the victim: ')
if money == 'yes':
    suma += 1

total = suma    
if total == 5:
    public = 'Murderers'
elif (total == 4) or (total == 3):
    public = 'Accomplices'
elif total == 2:
    public = 'Suspect'
elif total == 1:
    public = 'Released'
print(public)
    
    
```

```python
def investigar():
    cont = 0
    result = ['Inocente','Inocente','Suspeito','Cmplice','Cmplice','Assassino']
    perguntas =  ['Mora perto da vtima?',
              'J trabalhou com a vtima?',
              'Telefonou para a vtima?',
              'Esteve no local do crime?',
              'Devia para a vtima?']

    print('Responda as perguntas com Sim ou No: S/N')
    for item in perguntas:
        resp = input(f'1 - {item}')
        if resp in ['sim','SIM','S','s']:
            cont += 1
    print(f'Voc foi considerado: {result[cont]}')

investigar()
```

Faa um programa que olhe todos os itens de uma lista e diga quantos deles
so pares.

```python
def even_number(numbers):
    total = 0
    for i in numbers:
        if i%2 == 0:
            total += 1
    return total

lista = list(range(1,101))    
even_number(lista)
```

Faa uma funo que recebe duas listas e retorna a soma item a item dessas
listas.
Exemplo: Se a funo receber as listas [1,4,3] e [3,5,1], ento a funo deve
retornar [1+3, 4+5, 3+1] = [4, 9, 4].

```python
def join_lista(a,b):
    new_list = list()
    for i in range(len(a)):
        l_3 = a[i] + b[i]
        new_list.append(l_3)
    return new_list

l_1 = [1,2,3]
l_2 = [3,4,5]
join_lista(l_1,l_2)    
```
